**10. Implement the Predator-prey model in python**

Suppose fish starts at 10 hundreds, bears starts at 1 hundred

alpha = 1.1

delta = 0.1

gamma = 0.4

draw the plots(bear-time, fish-time)

t = np.linspace(0,50,num=1000)


```python
#https://scientific-python.readthedocs.io/en/latest/notebooks_rst/3_Ordinary_Differential_Equations/02_Examples/Lotka_Volterra_model.html
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import integrate
import ipywidgets as ipw

alpha = 1.1
beta = 0.4
delta = 0.1
gamma = 0.4
x0 = 0.0010
y0 = 0.001

def draw_the_plots(X, t, alpha, beta, delta, gamma):
    x, y = X
    dotx = x * (alpha - beta * y)
    doty = y * (-delta + gamma * x)
    return np.array([dotx, doty])
  
#num as Nt
Nt = 1000
#tmax as 50
tmax = 50
t = np.linspace(0,tmax, Nt)
X0 = [x0, y0]
res = integrate.odeint(draw_the_plots, X0, t, args = (alpha, beta, delta, gamma))
x, y = res.T

#I can print(x,y) and get results. Not sure what I was doing wrong.
print(x,y)


#I tried to use this portion, but it didn't show --
# I got this from the link in Modules, Lecture

plt.figure()
plt.grid()
plt.title("odeint method")
plt.plot(t, x, 'xb', label = 'Fish')
plt.plot(t, y, '+r', label = "Bear")
plt.xlabel('Time t, [days]')
plt.ylabel('Population')
plt.legend()

plt.show()



```
