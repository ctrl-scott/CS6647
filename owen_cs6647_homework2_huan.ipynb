10. Write an interactive Python program to play a “guess the  number” game. The computer should think of a random integer between  1 and 20 and the user (player) has to try to guess the number within six  attempts. The program should let the player know if the guess is too high or  too low. 


```python
import random
import string
import math

guessNum = random.randrange(1,21)

guessCount = 0

guessAllow = 6

while guessCount < guessAllow:
  guessUser = int(input("Enter your number guess (1 to 20): "))
  print("Your guess is: ", guessUser)
  if guessUser == guessNum:
    print("Congratulations! You have guessed the number! ", guessNum)
    break
  elif guessUser > guessNum:
      print("Your number guess was to high! try again!")
  elif guessUser < guessNum:
      print("Your number guess was to low! try again!")

      
```
