```python
import numpy as np
from sympy import sin,cos,symbols,integrate
import matplotlib.pyplot as plt
#modified from https://people.duke.edu/~ccc14/sta-663-2020/notebooks/S14D_Monte_Carlo_Integration.html



def f(x):
  return x * np.cos(71*x) + np.sin(13*x)

x = np.linspace(0,1,100)
plt.plot(x,f(x))
pass

x = symbols('x')
print("Integral Value\n------------------------\n", integrate(x * cos(71*x) + sin(13*x), (x, 0,1)).evalf(6))
print("\n")
print("Single MC Integration Estimate")
print("------------------------------")

n = 1000
x = f(np.random.random(n))
y = 1.0/n * np.sum(x)
y

print(y,"\n")
```
